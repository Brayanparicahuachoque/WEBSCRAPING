```python
productos = []

for product_lists in product_list:
    product_name = product_lists.find('a', class_='product-name').text.strip()
    product_price = product_lists.find('span', class_='price product-price').text.replace(' ', '').replace('₡','').strip()

    
    producto_actual = {'Nombre': product_name, 'Precio': product_price}

    
    productos.append(producto_actual)


df_productos = pd.DataFrame(productos)

print(df_productos)
```

```python
import matplotlib.pyplot as plt

plt.bar(df_productos['Nombre'], df_productos['Precio'])


plt.title('Precios de los productos')
plt.xlabel('Producto')
plt.ylabel('Precio')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.barh(df_productos['Nombre'], df_productos['Precio'])

plt.title('Precios de los productos')
plt.xlabel('Precio')
plt.ylabel('Producto')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Crear el gráfico de pastel
plt.pie(df_productos['Precio'], labels=df_productos['Nombre'], autopct='%1.1f%%')

# Personalizar el gráfico
plt.title('Distribución de precios de los productos')

# Mostrar el gráfico
plt.show()
```
